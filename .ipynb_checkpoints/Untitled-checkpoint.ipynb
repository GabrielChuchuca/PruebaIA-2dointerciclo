```python
# Busqueda en Amplitud - Breadth First Search

# Creamos la clase Nodo
class Node:
    def __init__(self, data, child=None): # Constructor de la clase
        self.data = data
        self.child = None
        self.fathr = None
        self.cost = None
        self.set_child(child)

    def set_child(self, child): # Agregar hijos
        self.child = child
        if self.child is not None:
            for ch in self.child:
                ch.fathr = self
            
    def equal(self, node): # Igual al equals de Java
        if self.data == node.data:
            return True
        else:
            return False

    def on_list(self, node_list): # Verfiicar su el nodo esta en la lista
        listed = False
        for n in node_list:
            if self.equal(n):
                listed = True
        return listed

    def __str__(self): # Igual al toString Java
        return str(self.data)
```

```python
# Implementacion del metodo de busqueda por amplitud
def search_Amplitud_solution(connections, init_state, solution):
    solved = False # Variable para almacenar el estado de la busqueda
    visited_nodes = [] # Nodos visitados
    frontrs_nodes = [] # Nodos en busqueda o lista nodos

    init_node = Node(init_state) # Nodo inicial
    frontrs_nodes.append(init_node)
    while (not solved) and len(frontrs_nodes) != 0:
        node = frontrs_nodes[0]
        # extraer nodo y añadirlo a visitados
        visited_nodes.append(frontrs_nodes.pop(0))
        if node.data == solution: # Preguntar se el nodo obtenido es la solucion
            solved = True
            return node # Retornamos el nodo de la solucion
        else:
            # expandir nodos hijo - ciudades con conexion
            node_data = node.data
            child_list = []
            for chld in connections[node_data]:
                child = Node(chld)
                child_list.append(child)
                if not child.on_list(visited_nodes) and not child.on_list(frontrs_nodes):
                    frontrs_nodes.append(child)
            node.set_child(child_list)

if __name__ == "__main__":
    connections = {
        'Iglesia_Central_La_Ascension_del_Señor': {'HOTEL_KAMARO', 'Parque_Zaracay', 'Hotel_Del_Pacifico'},
        'Cerro_Bombolí': {},
        'Parque_Zaracay': {'Restaurant_D_MARCO', 'Grand_Hotel_Santo_Domingo', 'Manduca_Wings_Grill'},
        'Jardín_Botánico_Padre_Julio_Marrero': {'Valle_Hermoso', 'San_Jacinto_del_Búa'},
        'San_Jacinto_del_Búa': {},
        'Valle_Hermoso': {},
        'Monumento_Del_Indio_Colorado': {'Victoria_Suites_Hostal', 'The_Craft_Cocina_Bar'},
        'Redondel_Sueño_de_Bolivar': {'Melkoke_Restaurant', 'Hotel_Santander_Don_Garcia'},
        'Hotel_Zaracay': {'Sabores_Grill_Latin_Bistro'},
        'Hostería_Samawa': {'Tinalandia_Lodge'},
        'Grand_Hotel_Santo_Domingo': {'El_Rincon_del_Che'},
        'Victoria_Suites_Hostal': {'Grand_Hotel_Santander'},
        'HOTEL_GOLDEN_VISTA': {'Redondel_Sueño_de_Bolivar'},
        'Hosteria_Kasadasa': {'Parque_Acuático_el_Pulpo_en_Santo_Domingo_D_Disney'},
        'Grand_Hotel_Santander': {},
        'Hotel_Puerta_Del_Sol': {'Monumento_Del_Indio_Colorado', 'Santo_Moro_Grill'},
        'Hotel_Santander_Don_Garcia': {'Jardín_Botánico_Padre_Julio_Marrero'},
        'Hotel_Toachi': {'Parrilladas_Che_Luis', 'Hotel_Zaracay'},
        'HOTEL_KAMARO': {'Redondel_Sueño_de_Bolivar'},
        'Hotel_Del_Pacifico': {'Conchal_Chabelita'},
        'Juan_Camarón_Santo_Domingo': {},
        'Casa_Bambú_Restaurante_Bar': {},
        'Sabores_Grill_Latin_Bistro': {'RUBGY_GRILL_BURGUER_SPORT'},
        'Conchal_Chabelita': {'Hotel_Puerta_Del_Sol'},
        'Santo_Moro_Grill': {},
        'Senior': {'HOTEL_GOLDEN_VISTA'},
        'El_Rincon_del_Che': {},
        'Señor_Bolon_Santo_Domingo': {},
        'Manduca_Wings_Grill': {},
        'Alitas_Del_Cadillac_SD': {'Club_del_Campo_Hotel_Centro_de_Convenciones'},
        'RUBGY_GRILL_BURGUER_SPORT': {'Grill_Fest', 'Juan_Camarón_Santo_Domingo', 'Chorigol_Santo_Domingo'},
        'Parrilladas_Che_Luis': {},
        'The_Craft_Cocina_Bar': {'Señor_Bolon_Santo_Domingo'},
        'Melkoke_Restaurant': {'Cerro_Bombolí', 'Casa_Bambú_Restaurante_Bar'},
        'La_Cevicheria_marisquerias_y_restaurantes_en_Santo_Domingo': {'Hostería_Samawa', 'Hosteria_Kasadasa'},
        'Agachaditos_Santo_Domingo': {'Hotel_Toachi'},
        'Restaurant_D_MARCO': {'Agachaditos_Santo_Domingo'},
        'Grill_Fest': {},
        'La_Cocina_De_Consuelo_Gourmet': {'La_Cevicheria_marisquerias_y_restaurantes_en_Santo_Domingo'},
        'Chorigol_Santo_Domingo': {'Alitas_Del_Cadillac_SD'}
    }

    init_state = 'Iglesia_Central_La_Ascension_del_Señor'
    solution = 'Alitas_Del_Cadillac_SD'
    solution_node = search_Amplitud_solution(connections, init_state, solution)
    # mostrar resultado
    result = []
    node = solution_node
    if node is not None: 
        while node.fathr is not None:
            result.append(node.data)
            node = node.fathr
        result.append(init_state)
        result.reverse() # Reverso el resultado (Solo para presentar)
        print(result)
    else:
        print("No hay solucion !!!!")
```
